```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time
from urllib.parse import quote
from datetime import datetime
```

```python
flipkart_website = 'https://www.flipkart.com/search?q=samsung+5g+mobile&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_8_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_8_na_na_na&as-pos=3&as-type=RECENT&suggestionId=samsung+5g+mobile&requestId=1216d363-e6fa-436d-91e4-dca986e985ce&as-backfill=on'
```

```python
amazon_website = 'https://www.amazon.in/s?k=samsung+5g+mobile&crid=2D3P4J64EDH32&sprefix=%2Caps%2C220&ref=nb_sb_ss_recent_1_0_recent' 
```

```python
headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'
        }
```

```python
webpage_flipkart = requests.get(flipkart_website, headers=headers) 
```

```python
webpage_amazon  =  requests.get(amazon_website, headers=headers) 
```

```python
# amazon links to products 
amazon_list = []
soup_amazon = BeautifulSoup(webpage_amazon.content, "html.parser")
a_links = soup_amazon.find_all("a", class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal")
 
for link in a_links:
        absolute_url = link.get('href')
        full_url = urljoin('https://www.amazon.in', absolute_url)
        amazon_list.append(full_url)
        time.sleep(1)
    
```

```python
# flipkart links to all products 
flipkart_list =[]
soup_flipkart = BeautifulSoup(webpage_flipkart.content,'html.parser')
f_links = soup_flipkart.find_all('a', class_='CGtC98')

for link in f_links:
        absolute_url = link.get('href')
        full_url = urljoin('https://www.flipkart.com', absolute_url)
        flipkart_list.append(full_url)
        time.sleep(1)
```

```python
# defining functions for both amazon
def get_title_amazon(soup_amazon):
    try:
        title=soup_amazon.find('span',{'id':'productTitle'}).get_text(strip=True)
        if title:
            return title
    except AttributeError:
        pass
    return 'NA'

def get_ratings_amazon(soup_amazon):
    try:
        ratings = soup_amazon.find('span', {'class': 'a-icon-alt'}).get_text(strip=True)
        if ratings:
            rating_value = re.search(r'\d+(\.\d+)?', ratings)
            if rating_value:
                return float(rating_value.group())
    except AttributeError:
        pass
    return None

def get_price_amazon(soup_amazon):
    try:
        price = soup_amazon.find('span',{'class':'a-price-whole'}).get_text(strip=True)
        if price:
            return price 
    except AttributeError:
        pass
    return 'NA'

def get_availability(soup_amazon):
    try:
        available = soup_amazon.find('div',{'id':'availability'}).get_text(strip=True)
        if available:
            return available
    except AttributeError:
        pass
    return 'NA'
```

```python
# defining functions for flipkart 
def get_availability_flipkart(soup_flipkart):
    try:
        available = soup_flipkart.find('div',{'id':'availability'}).get_text(strip=True)
        if available:
            return available
    except AttributeError:
        pass
    return 'In stock'
def get_price_flipkart(soup_flipkart):
    try:
        price = soup_flipkart.find('div',{'class':'Nx9bqj CxhGGd'}).get_text(strip=True)
        if price:
            return price
    except AttributeError:
        pass
    return 'NA'

def get_ratings_flipkart(soup_flipkart):
    try:
        ratings = soup_flipkart.find('div',{'class':'XQDdHH'}).get_text(strip=True)
        if ratings:
            return ratings
    except AttributeError:
        pass
    return 'NA'
def get_title_flipkart(soup_flipkart):
    try:
        title = soup_flipkart.find('span',{'class':'VU-ZEz'}).get_text(strip=True)
        if title:
            return title
    except AttributeError:
        pass
    return 'NA'
```

```python
amazon_data = {'Product_Name': [], 'Price': [], 'Ratings': [], 'Availability': []}
for link in amazon_list:
        new_page = requests.get(link, headers=headers)
        new_soup = BeautifulSoup(new_page.content, "html.parser")
        Product_Name = get_title_amazon(new_soup)
        Price = get_price_amazon(new_soup)
        Ratings = get_ratings_amazon(new_soup)
        Availability = get_availability(new_soup)
        
        amazon_data['Product_Name'].append(Product_Name)
        amazon_data['Price'].append(Price)
        amazon_data['Ratings'].append(Ratings)
        amazon_data['Availability'].append(Availability)
        time.sleep(1)
        
website_df = pd.DataFrame.from_dict(amazon_data)
website_df.to_csv("Amazon_Website_data.csv", header=True, index=False)
```

```python
flipkart_data = {'Product_Name': [], 'Price': [], 'Ratings': [], 'Availability': []}
for link in flipkart_list:
        new_page = requests.get(link, headers=headers)
        new_soup = BeautifulSoup(new_page.content, "html.parser")
        Product_Name = get_title_flipkart(new_soup)
        Price = get_price_flipkart(new_soup)
        Ratings = get_ratings_flipkart(new_soup)
        Availability = get_availability_flipkart(new_soup)
        
        flipkart_data['Product_Name'].append(Product_Name)
        flipkart_data['Price'].append(Price)
        flipkart_data['Ratings'].append(Ratings)
        flipkart_data['Availability'].append(Availability)
        time.sleep(1)
        
        
flipkart_df = pd.DataFrame.from_dict(flipkart_data)
flipkart_df.to_csv("Flipkart_Website_data.csv", header=True, index=False)  
```
